**Objective:** To build a machine learning model that predicts the presence of heart disease based on patient clinical features.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Step 2: Load Dataset
df = pd.read_csv('/kaggle/input/heart-disease-prediction/Heart_Disease_Prediction.csv')
df.head()
```

```python
# Step 3: Explore the Data
print(df.info())
print(df.describe())
print(df.columns)
print(df.isnull().sum())
```

```python
# Step 4: Check Unique Values in Target Column
# (Figure out which column is the target by inspecting the data)
print(df['Heart Disease'].value_counts())  # This is likely the target column
```

```python
# Step 5: Encode Categorical Columns (if any)
# Check for object types
print(df.dtypes)

# Example: Encode 'Sex', 'Chest Pain Type', etc.
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
print(df_encoded.columns.tolist())
```

```python
print(df.columns)
```

```python
# Separate the target variable
y = df['Heart Disease']

# Drop the target from the feature set
X = df.drop('Heart Disease', axis=1)
```

```python
# Use get_dummies to convert categorical columns
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```
